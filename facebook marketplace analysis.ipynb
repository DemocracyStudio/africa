# Facebook marketplace analysis in five African cities
- Cape Town, South Africa
- Lomé, Togo
- Yaoundé, Cameroon
- Nairobi, Kenya
- Addis Ababa, Ethiopia

### 1. Scrape marketplace items
- Get description, link to post, imgurl
- For each post, get the full seller's comment.

### 2. Feature extraction
- product classification/categorization using description
- objects recognition using images 
- extract location, addresses, using Named Entity Recognition
- get lat/long from addresses extracted

### 3. Exploratory data analysis
- which are the most sold categories of products in each city?
- which are the frequency of offers?
- Are there geographical clusters of products? ex. food in some districts, clothes in another...
- Which products are proposed for Free?
- What are the average price for each category of product?

### 4. Predictive model with Machine Learning
- Create a price suggestion based on category of product and location.

# Cape Town marketplace

```python
# start the browser
driver = webdriver.Chrome(executable_path = '/usr/local/bin/chromedriver')

# open Cape Town marketplace
url = 'https://web.facebook.com/marketplace/capetown/search/?query=cape%20town&exact=false'
driver.get(url)
```

```python
# scroll down automatically
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    # Wait to load page
    time.sleep(randint(1,5))
    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
# scrape elements, adding a random sleep
elements = driver.find_elements_by_class_name('sonix8o1')
# create empty list
descriptions = []
imgurls = []
links=[]
#count=0

# get title, price, description
for i in elements:
    #print(count)
    #print(i.get_attribute('title'))
    print(i.text)
    descriptions.append(i.text)
    links.append(i.get_attribute('href'))
    #display(HTML(i.get_attribute('outerHTML'))) #display pictures
    imgurls.append(i.get_attribute('outerHTML'))
    print('-----------')
    #count+=1
    time.sleep(random.uniform(0.1,0.9))
```

```python
print(len(descriptions))
print(len(links))
print(len(imgurls))
```

```python
# when length match:
# build a dataframe with data collected
capetown = pd.DataFrame(list(zip(descriptions, links, imgurls)),
                        columns =['description', 'link', 'imgurl'])
capetown
```
